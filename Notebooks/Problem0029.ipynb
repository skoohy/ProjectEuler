# Problem 29: Distinct powers

### Description:

Consider all integer combinations of $a^b$ for $2 ≤ a ≤ 5$ and $2 ≤ b ≤ 5$:

$$2^2=4, \quad 2^3=8, \quad 2^4=16, \quad 2^5=32$$
$$3^2=9, \quad 3^3=27,\quad 3^4=81, \quad 3^5=243$$
$$4^2=16, \quad 4^3=64,\quad 4^4=256,\quad 4^5=1024$$
$$5^2=25,\quad 5^3=125,\quad 5^4=625,\quad 5^5=3125$$
If they are then placed in numerical order, with any repeats removed, we get the following sequence of 15 distinct terms:

$$4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125$$

How many distinct terms are in the sequence generated by $a^b$ for $2 ≤ a ≤ 100$ and $2 ≤ b ≤ 100$?

This will be a very quick and simple solution. We start with two for loops both for $a,b$ from $2$ to $100$ and appending $a^b$ to a list.

```python
powers = [a**b for a in range(2,101) for b in range(2,101)]
```

I will first present my initial solution which runs in less than 0.5 seconds, then present a one-liner that is approximately $98\%$ faster. 

We want to look for the powers that are distinct or rather find which ones are repeated and forget about them. We can create a new list containing the distinct tearms. After that we can create a loop checking if a value is not in our new list, if it isn't append it.

```python
import time
t1 = time.time()

powers = [a**b for a in range(2,101) for b in range(2,101)] # all terms
        
new = [] # new list containing distinct terms
for i in powers:
    if i not in new: # if i is not in new add it to new
        new.append(i)

print(len(new))

t2 = time.time()
print(f'Program Execution Time: {t2-t1} seconds')
```

We simplify this down, and make it much faster by applying the $set()$ function to the powers list and then applying the $len()$ function. The $set()$ function will exclude all repeated elements.

```python
import time
t1 = time.time()

print(len(set([a**b for a in range(2,101) for b in range(2,101)])))

t2 = time.time()
print(f'Program Execution Time: {t2-t1} seconds')
```
